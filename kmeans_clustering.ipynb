```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("D:/courses/[FreeCourseSite.com] Udemy - The Data Science Course Complete Data Science Bootcamp/38 - Advanced Statistical Methods - K-Means Clustering/002 3.01.Country-clusters.csv")
data
```

```python
plt.scatter(data["Longitude"], data["Latitude"])
plt.xlim(-180, 180)
plt.ylim(-90, 90)
plt.show()
```

Selecting Numerical Features:

```python
#we are gonna select longitude and latitude
x = data.iloc[:,1:3]
x
```

Clustering:

```python
kmeans = KMeans(3)
kmeans.fit(x)
```

```python
#clustering results:
identified_clusters = kmeans.fit_predict(x)
data_with_clusters = data.copy()
data_with_clusters["Cluster"] = identified_clusters
data_with_clusters
```

```python
plt.scatter(data_with_clusters["Longitude"], data_with_clusters["Latitude"], c = data_with_clusters["Cluster"], cmap = "rainbow")
plt.xlim(-180, 180)
plt.ylim(-90, 90)
plt.show()
```

Clustering Categorical Data:

```python
#now we will apply kmean clustering using language
#data mapping:
data_mapped = data.copy()
data_mapped["Language"] = data_mapped["Language"].map({"English": 0, "French": 1, "German": 2})
data_mapped
```

```python
#selecting features (language):
x = data_mapped.iloc[:,3:4]
x
```

```python
#clustering:
kmeans = KMeans(3)
kmeans.fit(x)
```

```python
#clustering results:
identified_clusters = kmeans.fit_predict(x)
data_with_clusters = data.copy()
data_with_clusters["Cluster"] = identified_clusters
data_with_clusters
```

```python
plt.scatter(data_with_clusters["Longitude"], data_with_clusters["Latitude"], c = data_with_clusters["Cluster"], cmap = "rainbow")
plt.xlim(-180, 180)
plt.ylim(-90, 90)
plt.show()
```
