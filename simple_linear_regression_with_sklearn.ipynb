```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("D:/courses/[FreeCourseSite.com] Udemy - The Data Science Course Complete Data Science Bootcamp/34 - Advanced Statistical Methods - Linear Regression with sklearn/003 1.01.Simple-linear-regression.csv")
```

```python
x = data["SAT"]
y = data["GPA"]
```

```python
x_matrix = x.values.reshape(-1, 1)  #
```

```python
reg = LinearRegression()
reg.fit(x_matrix, y)
```

```python
plt.scatter(x, y)
yhat = reg.coef_ * x_matrix + reg.intercept_
fig = plt.plot(x, yhat, c = "orange", label = "regression line")
plt.xlabel("SAT", fontsize = 20)
plt.ylabel("GPA", fontsize = 20)
plt.show()
```
